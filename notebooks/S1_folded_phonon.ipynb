# S1 折叠声子验证


```python
import numpy as np
import matplotlib.pyplot as plt

from src.experiments_simple import folded_phonon_demo
from src.visualization import setup_plot_style, save_figure

setup_plot_style()

n_cells = 40
r1 = 1 / 4
r2 = 3 / 8
m_supercell = 4

t0 = 1.0
alpha = 0.5

a = 1.0
mass = 1.0

q_branches, g_vals = folded_phonon_demo(
    n_cells=n_cells,
    r1=r1,
    r2=r2,
    m_supercell=m_supercell,
    t0=t0,
    alpha=alpha,
    a=a,
    mass=mass,
)

fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(np.arange(len(q_branches)), np.abs(g_vals))
ax.set_xlabel("折叠分支索引")
ax.set_ylabel(r"$|g(q)|$")
ax.set_title("折叠分支矩阵元对比")

save_figure(fig, "results/figures/S1_folded_phonon")
```
